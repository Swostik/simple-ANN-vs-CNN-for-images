# Convolution Neural Network for Image Classification

- Images were downloaded from https://www.kaggle.com/prasunroy/natural-images
- This dataset contains 6,899 images from 8 distinct classes compiled from various sources. The classes include airplane, car, cat, dog, flower, fruit, motorbike and person.

## Preprocessing 

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import tensorflow as tf
from sklearn.model_selection import train_test_split
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow.keras as keras
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#working directory
my_path = 'natural_images'
categories = os.listdir(my_path) #lists the files and folders inside the my_path
```

### Images to array

```python
#changing each image to its corresponding array values
data =[]
labels=[]
for category in categories:
    image_path = os.path.join(my_path,category)
    images = os.listdir(image_path)
    for image in images:
        img = os.path.join(image_path,image)
        img = load_img(img,target_size=(64,64))
        img=img_to_array(img)
        data.append(img)
        labels.append(category)
```

```python
data = np.array(data)
data.shape
```

### Integer Encoding with dictionary

```python
#creating dictionary to map the values with the labels ...integer encoding
dic={}
a= 0
for category in categories:
    dic[category]=a
    a += 1
    
dic   
```

```python
# changing into pandas series to map the values to the dictionary
new_labels = pd.Series(labels)
new_labels = new_labels.map(dic)
new_labels = np.array(new_labels)
new_labels
```

```python
#normalizing the images so that each value is between 0 and 1
X = data/255
```

### Train_Test Split

```python

X_train, X_test, y_train,y_test = train_test_split(X,new_labels)
print(f'The train set shape : {X_train.shape}')
print(f'The test set shape : {X_test.shape}')
```

## Model with just dense layers

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=[64,64,3]))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(512, activation='relu'))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(512, activation='relu'))

model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))


model.add(keras.layers.Dense(8, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs = 50, validation_split=0.3)
```

### Graphical view of training and validation loss and accuracy

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
```

```python
plt.subplots(1,2, figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(train_loss,label='Training loss')
plt.plot(val_loss,label='Validation loss')
plt.legend()
plt.subplot(1,2,2)
plt.plot(train_accuracy,label='Training accuracy')
plt.plot(val_accuracy,label='Validation accuracy')
plt.legend();
```

### Classification report

```python
y_pred = model.predict(X_test)
y_pred_classes = [np.argmax(pred) for pred in y_pred]
print(classification_report(y_test,y_pred_classes))
print(confusion_matrix(y_test,y_pred_classes))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred_classes), annot=True,fmt='d');
```

- This heatmap shows how the model is unable to identify most of the label 3(dog) as dog but rather it identifies it as label 2(cat).

### Images agains the predicted classes

```python
#switching keys and values of the dictionary so that it will easier to plot the images
labels = {}
for (k,v) in dic.items():
    labels[v] = k
print(labels)
```

```python
plt.subplots(nrows=3,ncols=3)
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.axis('off')
    plt.imshow(X_test[i])
    plt.title(labels[y_pred_classes[i]])
  
```

### Conclusion
- From the classficiation report and images that can be seen above, we can see how this model is not able to distinguish cat,dog, person. 

## Convolution layer

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(filters=10, kernel_size=(3,3), activation='relu',input_shape=(64,64,3)))
model.add(keras.layers.Conv2D(filters=10, kernel_size=(3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))

model.add(keras.layers.Flatten())


model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dense(64, activation='relu'))


model.add(keras.layers.Dense(8, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs = 20, validation_split=0.3)
```

### Graphical view of training and validation loss and accuracy

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
```

```python
plt.subplots(1,2, figsize=(10,6))

plt.subplot(1,2,1)
plt.plot(train_loss,label='Training loss')
plt.plot(val_loss,label='Validation loss')
plt.legend()
plt.subplot(1,2,2)
plt.plot(train_accuracy,label='Training accuracy')
plt.plot(val_accuracy,label='Validation accuracy')
plt.legend()
```

### Classification report

```python
y_pred = model.predict(X_test)
y_pred_classes = [np.argmax(pred) for pred in y_pred]
print(classification_report(y_test,y_pred_classes))
print(confusion_matrix(y_test,y_pred_classes))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred_classes), annot=True,fmt='d');
```

- In this case though as oppose to previous heatmap, it can classify more images as its correct labels

### Images against the predicted classes

```python

plt.subplots(nrows=3,ncols=3)
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.axis('off')
    plt.imshow(X_test[i])
    plt.title(labels[y_pred_classes[i]])
  
    
```

### Conclusion
- As we can see here how the convolution layer accurately predicts most of the images and the accuracy and support for each classes in the classification report are much higher than the model using just dense layers  

## Summary


Convolutional Neural Networks are very similar to ordinary Neural Networks : they are made up of neurons that have learnable weights and biases. Each neuron receives some inputs, performs a dot product and optionally follows it with a non-linearity. The whole network still expresses a single differentiable score function: from the raw image pixels on one end to class scores at the other. And they still have a loss function (e.g. SVM/Softmax) on the last (fully-connected) layer and all the tips/tricks we developed for learning regular Neural Networks still apply.

So what changes? ConvNet architectures make the explicit assumption that the inputs are images, which allows us to encode certain properties into the architecture. These then make the forward function more efficient to implement and vastly reduce the amount of parameters in the network.
(Reference: https://cs231n.github.io/convolutional-networks/)
